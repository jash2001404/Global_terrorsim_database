```python
import pandas as pd 
import plotly.express as px 
import numpy
import plotly.io as pio
pio.renderers.default = "browser"
  
df = pd.read_csv('gtdmate.csv')
fig = px.scatter_geo(data_frame=df, lat="latitude",lon="longitude", 
                    color="targtype1_txt",
#                      hover_name="country_txt",
                     animation_frame="iyear",
                      labels={"country_txt":"Country","iyear":"Year",
                              "targtype1_txt":"Type of Target",
                              "attacktype1_txt":"Type of Attack",
                              "natlty1_txt":"Nationality of Victims"},
                     projection="natural earth"
                     # projection="orthographic"
                     )


button_layer_1_height = 1.01
fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                    dict(
                        args=["color","targtype1_txt"],
                        label="Targets of Attack",
                        method="update"
                    ),dict(
                        args=["color","country_txt"],
                        label="Countries",
                        method="update"
                    ),dict(args=["color","targtype1_txt"],
                        label="Targets of Attack",
                        method="update"
                    ),dict(
                        args=["color","country_txt"],
                        label="Countries",
                        method="update"
                    ),dict(
                        args=["color","attacktype1_txt"],
                        label="Attack Type",
                        method="update"
                    ),dict(
                        args=["color","natlty1_txt"],
                        label="Nationality of Victims",
                        method="update"
                    )]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.001,
            xanchor="left",
            y=button_layer_1_height,
            yanchor="top"
            )
    ]
)
fig.update_layout(title_text="Terror events over the years based on different variables ")

fig

```

```python
import pandas as pd 
import plotly.express as px 
import numpy
import plotly.io as pio
pio.renderers.default = "browser"
  
df = pd.read_csv('gtdmate.csv')
fig = px.scatter_geo(data_frame=df, lat="latitude",lon="longitude", 
                    color="region_txt",
                     size=df["nkill"],size_max=100,
                     animation_frame="iyear",
                      labels={"country_txt":"Country","iyear":"Year",
                              "targtype1_txt":"Type of Target",
                              "attacktype1_txt":"Type of Attack",
                              "natlty1_txt":"Nationality of Victims","nkill":"Casualties","region_txt":"Region"},
                 projection="natural earth"
#                       projection="orthographic"
                     )
fig.update_layout(title_text="Casualities in terror events over the years")

fig
```
